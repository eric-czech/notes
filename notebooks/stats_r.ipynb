####Empirical Rule

```python
%%R
req('ggplot2')
emp_rule = function(u=10, s=5){
  x = seq(u - s * 4, u + s * 4, length.out = 100)
  y = dnorm(x, mean=u, sd=s)
  # Rule breaks as half-tails
  b = c( 0.0125, 0.025, 0.16, 0.5, 0.84, 0.975, 0.9875 )
  
  # Plot PDF
  plot = ggplot(data.frame(x=x, y=y), aes(x=x, y=y)) + geom_line() + theme(panel.grid.minor=element_blank())
  
  # Add annotations for rule thresholds
  for(i in b){
    v = qnorm(i, mean=u, sd=s)
    plot = plot + geom_vline(xintercept=v, linetype='longdash')
    if (i > .5){
      lbl = paste0(round((i - .5) * 200, 1), '%')
      plot = plot + geom_text(data=NULL, x=v, y=max(y)/2, vjust=-.5, label=lbl, angle=270) + 
        theme(panel.grid.major = element_blank(), panel.grid.minor = element_blank(), panel.background = element_blank())
    }
  }
  plot
}
emp_rule()
```

#### Permutations

```python
%%R
# Create a list, find all permutations of that list, and make 
# sure the number of permutations matches what is expected
req('gtools')
d = c(1,2,3)
n = length(d)
p = permutations(n=n, r=n, v=d)

to_list = function(p){
    paste0('[', paste(apply(p, 1, function(x){paste0('(', paste(x, collapse=',') ,')')}), collapse=', '), ']')
}

# Without replacement/repitition
print(sprintf('Permutations of %s = %s', paste(d, collapse=','), to_list(p)))
print(sprintf('Actual permutations length = %d, expected length = %d', nrow(p), factorial(n)))

# With 
p = permutations(n=n, r=n, v=d, repeats.allowed=T)
print(sprintf('Permutations of %s = %s', paste(d, collapse=','), to_list(p)))
print(sprintf('Actual permutations length = %d, expected length = %d', nrow(p), n ^ n))
```

#### Combinations

```python
%%R
# Show combinations of 3 from 1:5 with and without replacement
d = c(1,2,3,4,5)
n = length(d)
r = n - 2

nCr = function(n, r){
    factorial(n) / ( factorial(r) * factorial(n-r) )
}

# Without replacement/repitition
p = combinations(n=n, r=r, v=d)
print(sprintf('Permutations of %s = %s', paste(d, collapse=','), to_list(p)))
print(sprintf('Actual permutations length = %d, expected length = %d', nrow(p), nCr(n, r)))

# With repitition
p = combinations(n=n, r=r, v=d, repeats.allowed=T)
print(sprintf('Permutations of %s = %s', paste(d, collapse=','), to_list(p)))
print(sprintf('Actual permutations length = %d, expected length = %d', nrow(p), nCr(n+r-1, r)))
```

#### Binomial Tests

Assume we have 57 successes of 150 trials where p = .3 and want to test alternative that p is greater than .3.  Note in the following how the "alternative" includes the observed value so $p$ is the probability of seeing values _at least_ as extreme:

```python
%%R

# p is probability of seeing the observed value AND greater
p = c(
    # dbinom results
    1 - sum(dbinom(0:56, 150, .3)),
    sum(dbinom(57:150, 150, .3)),

    # pbinom results
    pbinom(56, 150, .3, lower.tail = F),
    1 - pbinom(56, 150, .3),

    # binom.test results
    binom.test(57, 150, .3, alternative='greater')$p.value
)
p # Ho rejected
```
