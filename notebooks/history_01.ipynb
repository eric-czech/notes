##<center>Hypothesis vs Significance Testing</center>

- Fisher's "Significance Testing" (seen for the first time is his 1925 publication 'Statistical Methods for Research Workers')
    - Says that a sample should determine the "significance" of a sample as a p-value
    - Said that p-values should be treated as a continuous measure of significance and it was is 1925 publication that originally proposed the 5% signifiance level
    - Said that null hypotheses are never accepted because a failure to reject does not prove that a point estimate for the null hypothesis is the true one (i.e. there may still be a difference of practical significance)
- Neyman + Pearson "Hypothesis Testing" (seen in a series of papers by the two starting in 1933)
    - Said that having to pick a significance level was a bad thing and instead that the smallest difference of practical significance should be used as an __alternative__ hypothesis

    - 1938 - Created Wilks's Theorem stating that $-2 ln( \Delta )$ where $\Delta$ is ratio of likelihoods is a $\chi^2$ variable

    - Created martingales in 1934
    - Created martingales as a part of a proof of the central limit theorem, stating that there were two conditions necessary for a sequence of random numbers to be normally distributed:
        1. The variability in the values is bounded 
        2. The best estimate of the next number is the current number
    - Applied by __Odd Aalen__ to clinical trials in the 1970s (said course of patient responses was a martingale)
