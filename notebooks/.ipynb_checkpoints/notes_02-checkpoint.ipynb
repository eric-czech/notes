## Types of Missingness
- Notation used:
    - $I$ is missingness indicator
    - $\theta$ is parameters for data distribution
    - $\phi$ is parameters for missing data process
- MAR - missing at random
    - A univariate data set $y$ is MAR if $p(I|y_{obs}, y_{miss}, \phi) = p(I|y_{obs}, \phi)$
    - The missingness process is _ignorable_ if $\theta$ and $\phi$ are independent, which more practically means that they don't have a joint prior distribution
- MCAR - missing completely at random
    - A data set $y$ is MCAR if $p(I|y_{obs}, y_{miss}, \phi) = p(I|\phi)$
- MNAR - missing not at random
    - Data is MNAR if the missingness is process is dependent on unobserved values
        - For example, suppose a survey asks about level education and earnings -- if people with higher levels of education are less likely to respond to the earnings question AND people sometimes fail to respond about their level of education, then earnings is MNAR

## Multiple Imputation

- __Joint MI__ - Treat all variables for a case as having come from a _joint_ multivariate normal distribution.
    - Examples in R are Amelia and Norm
    - Assumes each case is a single multivariate normal draw
- __Conditional MI__ - Use observed variables to predict existing ones by drawing from a conditional distribution
    - Examples in R are mi and mice
    
- Joint MI is less accurate than conditional MI when using categorical variables (see pg 2 of Resource 1)
- Joint and Conditional MI are similar in accuracy on continous data (though Joint can often be slightly better)
- Joint MI runs much more quickly than Conditional

Notes on R packages:

- Amelia does not work if there are 0 complete cases and it complains if there is 1 __column__ (not case) having only 1 or 0 non-na values
- Mice seems to silently impute NA values if variables are highly collinear
    
## Joint Modeling

### Maximum Likelihood Estimates
See resource #2

In this approach, data is assumed to be multivariate normal and the likelihood for each record is the likelihood of the observed variables integrated over the likelihood of all possible missing variables.

For example, if two observations were missing from a case with $k$ total variables, the likelihood is:

$$ L(X|\Theta) = \int{L(x_1, x_2, ... x_{k-2}|\Theta) dx_{k-1} dx_k }$$

The likelihood of the whole dataset is then the product of the above for each case.  $\Theta$ is maximized over the joint likelihood.

##<center>Proportional Hazards</center>

### Estimating a Survival Function from Cox Regression Model

(taken from [here (page 3)](http://www.jstatsoft.org/v61/c01/paper))
- Survival Function: $ S(t|X) = e^{-\Lambda(t|X)} $ where $\Lambda(t|X)$ is the cumulative hazard
- In the cox PH model, the cumulative hazard which normally is defined as $\int_0^t{\lambda(u|X)du} $, simplifies nicely to:
    $$ e^{\beta X}\Lambda_0(t)$$ where $\Lambda_0(t)$ is the cumulative baseline hazard
- $\Lambda_0$ then has an estimator called $\hat{\Lambda_0}$ (see the paper above on pg 4 for it) based on the data itself, and that estimator can be plugged back into the Survival Function to give:
    $$ \hat{S}(t|x^*) = e^{-e^{\beta X}\hat{\Lambda_0}(t)}$$ where $x^*$ is a single setting for the covariates

##<center>Expectation Maximization</center>

# Gaussian Mixtures

Source code examples:
1. [C++ Example 1](http://csg.sph.umich.edu/abecasis/class/2006/615.18.pdf)

#### Assumptions
- Assume probability of mixture $Z$ is $\pi_z, z \in {1, 2, ..., m}$
- Assume mixture means is $\theta_z, z \in {1, 2, ..., m}$
- Assume $n$ data points $x_i, i \in {1, 2, ..., n}$

#### Algorithm
1. Compute $P(Z=z|x_i)$ for each $i$ -- In other words, get the likelihood of each mixture, for each data point
2. Set the new mixture proportions as the sum of the above probability for each mixture number
    $$P(Z=z) = \sum_{i=1}^n{P(Z=z|x_i)}$$
3. Set the new mixture means as the sum of the $P(Z=z|x_i)$ for each point times its value, $x_i$.  Divide this sum by the number of points overall, $n$, times the mixture probability for $z$, $\pi_z$
    $$\theta_z = \frac{\sum_{i=1}^n{x_i \cdot P(Z=z|x_i)}}{n \cdot \pi_z}$$

##<center>Classifier Scoring</center>

- AUC for precision-recall is better than ROC if problem is imbalanced
    - This is because the precision and recall scores do not include __true negatives__ (TN)
    - TN is not important when the number of positive cases is small
- Recall/Sensitivity is important for things like detecting cancer when you want a true positive to be predicted positively even if it's at the expense of precision, in the form of false-positives


## Classifier Calibration

- Boosted regression trees and SVM's are known to cluster predictions near .5 (forming a sigmoid in the reliability diagram)
- Calibration takes care of this, and is often viewed through __reliability diagrams__
    - These diagrams work by showing the fraction of positive cases vs the probability predicted (they should match up to create a straight line)
- Platt Scaling and Isotonic Regression are both ways to calibrate with isotonic regression generally being preferred

##<center>Regression To The Mean</center>

- Defined by Sir Francis Galton in his 1886 paper 'Regression towards mediocrity in hereditary stature'
- States that for a bivariate distribution with __exactly equal__ marginal distributions (so far I've only seen 2-dimensional multivariate guassian apply), the expected value of one observation given the other lies closer to it's mean 
    - Basically, if $X$ and $Y$ are 2-dim multivariate normal, then:
        1. They must have equal variance to have the same marginals
        2. $\hat{Y} <= E(Y|X=x) < x$ -> This is the essence of regression to the mean
- Galton's example for this, the first example of it, came from looking at the distribution of children vs their parents.  The distribution of children overall follows the same distribution as the parents (i.e. the marginals are equal), yet the average height of children still ended up being more towards the mean overall rather than being centered on the parents height

##<center>Noninformative Priors vs Frequentist Methods</center>

- For single parameter estimation, there seem to be no examples with major differences (yet)

### [Jake Vanderplas Article](http://jakevdp.github.io/blog/2014/06/06/frequentism-and-bayesianism-2-when-results-differ/)

Major differences in practice between bayesian and frequentist methods:
1. In the handling of nuisance parameters (frequentist methods would use a point estimate (ML estimate) while bayesian methods would marginalize -- bayesian methods agree with monte carlo simulation of the same scenario)
2. In the differences between credible intervals and confidence intervals


- Some notes on the differences:
    - Bayesianism treats parameters (e.g. μ) as random variables, while frequentism treats parameters as fixed.
    - Bayesianism treats observed data (e.g. D) as fixed, while frequentism treats data as random variables.
    - Bayesianism treats its parameter constraints (e.g. CRμ) as fixed, while frequentism treats its constraints (e.g. CIμ) as random variables.

##<center>Centering Predictors For Interactions</center>

- Assume you have two predictors $X_1$ and $X_2$ where $X_{1_c}$ is the centered version of that predictor
- A regression of $Y$ on $X_1$ and $X_2$ will produce the same coefficients as a regression on $X_{1_c}$ and $X_{2_c}$
- The coefficients for a regression on $X_1$, $X_2$, and $X_1 : X_2$ will be different for $X_1$ and $X_2$ as the regression without an interaction
- However, if the centered variables are used instead then the coefficients for $X_{1_c}$ and $X_{2_c}$ will be "similar" even if the interaction is included

This is a __crucial__ difference when it comes to averaging parameter estimates across models.

- Note that factors, binary, and continuous variables should all be centered to do model averaging

Source: http://seananderson.ca/2014/07/30/centering-interactions.html

##<center>Censoring vs Truncation</center>

- __Censoring__ - Results in observations that are capped either from above or below
- __Truncation__ - Results in certain observations not even being collected due to some cap in the collection process

See [this SO post](http://stats.stackexchange.com/questions/144041/what-is-the-difference-between-censoring-and-truncation) for an example.
