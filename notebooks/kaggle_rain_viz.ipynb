```python
def scatter_plot(data, x, y):
    data = data[[x, y]]
    data = data.loc[np.any(data > 0, axis=1)]
    data.plot(kind='scatter', x=x, y=y)
    
def hexbin_plot(data, x, y):
    data = data[[x, y]]
    data = data.loc[np.all(data >= 0, axis=1)]
    plt.hexbin()
```

```python
from pylfer import utilities
#cols = ['ReflectivityQC', 'Reflectivity', 'Zdr', 'LogWaterVolume', 'Expected']
sentinels = {
    0: 'zero',
    -99900: 'not-sure',
    -99000: 'echo-low',
    -99901: 'range-folded',
    -99903: 'no-collect',
    999.0: 'echo-edge'
}
#cols = ['RhoHV', 'RadarQualityIndex', 'Expected']
cols = ['HybridScan', 'Zdr', 'Expected']
d = data[cols]
d['Expected'] = d['Expected'].apply(lambda x: math.asinh(x))
hists = utilities.get_histograms(d, 'Expected', sentinels=sentinels, bins=12)
plots = utilities.plot_histograms(hists, cnorm=matplotlib.colors.LogNorm(), cmap=matplotlib.cm.ocean_r)
for plot in plots:
    plot[0].set_size_inches((12,12))
```

```python
idx_cols = ['IdIdx', 'ScanIdx']
ext_cols = ['TimeToEnd', 'DistanceToRadar']
ind_cols = [c for c in data.columns if c not in ['Id', 'Expected'] + idx_cols + ext_cols]
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Imputer

clf = Pipeline([
    ('impute', Imputer(strategy='median')),
    ('clf', ExtraTreesRegressor())
])
X, y = data[ind_cols], data['Expected']
clf.fit(X, y)
```

```python
pd.Series(clf.named_steps['clf'].feature_importances_, index=X.columns).order().plot(kind='bar')
```

```python
from sklearn.datasets import load_iris
from scipy.stats import f_oneway
dt = load_iris()
d = pd.DataFrame(dt.data, columns=dt.feature_names)
d['class'] = pd.Series(dt.target).map(dict([(i, v) for i, v in enumerate(dt.target_names)]))
target = 'sepal length (cm)'
datasets = [ d[d['class'] == c][target] for c in dt.target_names ]
f_oneway(*datasets)
```
